```python
import pandas as pd
import matplotlib.pyplot as plt

seattle_weather = pd.read_csv('C:/Users/nswat/Documents/DataCamp/Datasets/5IntroToDataVisMatplotlib/seattle_weather3.csv', 
                          index_col=0)
austin_weather = pd.read_csv('C:/Users/nswat/Documents/DataCamp/Datasets/5IntroToDataVisMatplotlib/austin_weather3.csv', 
                              index_col=0)

fig, ax = plt.subplots()

ax.errorbar(seattle_weather['MONTH'], 
            seattle_weather['MLY-TAVG-NORMAL'], 
            yerr=seattle_weather['MLY-TAVG-STDDEV'])

ax.errorbar(austin_weather['MONTH'], 
            austin_weather['MLY-TAVG-NORMAL'], 
            yerr=austin_weather['MLY-TAVG-STDDEV'])

ax.set_ylabel('Temperature (Fahrenheit)')
plt.show()
```
