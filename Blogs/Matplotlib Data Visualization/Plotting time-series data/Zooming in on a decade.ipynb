```python
import pandas as pd
import matplotlib.pyplot as plt

climate_change = pd.read_csv('C:/Users/nswat/Documents/DataCamp/Datasets/5IntroToDataVisMatplotlib/climate_change.csv')

# First, convert the column (it's a string) to datetime type
datetime_series = pd.to_datetime(climate_change['date'])

# Then, create datetime index passing the datetime series
datetime_index = pd.DatetimeIndex(datetime_series.values)

# New climate change DataFrame with datetime type for date column
df2=climate_change.set_index(datetime_index)

# We don't need the date column anymore since the years are now the DataFrame's index
df2.drop('date',axis=1,inplace=True)

#Define the 1960s decade
sixties = df2['1960-01-01':'1969-12-31']

#Create a graph that zooms in on the 1960s decade
fig, ax = plt.subplots()
ax.plot(sixties.index, sixties['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
```
