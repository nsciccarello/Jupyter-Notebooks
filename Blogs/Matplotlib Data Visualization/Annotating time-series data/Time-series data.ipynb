```python
import pandas as pd
import matplotlib.pyplot as plt

climate_change = pd.read_csv('C:/Users/nswat/Documents/DataCamp/Datasets/5IntroToDataVisMatplotlib/climate_change.csv',
                            parse_dates=['date'],
                            index_col='date')

# plot_timeseries function:
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)

fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, 
                climate_change['co2'],
                'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index,
                climate_change['relative_temp'],
                'red', 'Time', 'Relative temperature (Celsius)')

plt.show()
```
