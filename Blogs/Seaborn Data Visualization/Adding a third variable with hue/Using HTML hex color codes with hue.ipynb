```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')

hue_colors = {'Yes':'#808080', 
              'No':'#00FF00'}

sns.scatterplot(x='total_bill',
                y='tip',
                data=tips,
                hue='smoker', 
                palette=hue_colors)

plt.show()
```
