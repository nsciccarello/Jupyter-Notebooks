```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

gdp_data = pd.read_csv('C:/Users/nswat/Documents/DataCamp/Datasets/6SeabornDataVisualization/countries-of-the-world.csv')

g = sns.catplot(x='Region', 
                y='Birthrate', 
                data=gdp_data,
                kind='box')

g.set(xlabel='New X Label', 
      ylabel='New Y Label')

plt.xticks(rotation=90)
plt.show()
```
