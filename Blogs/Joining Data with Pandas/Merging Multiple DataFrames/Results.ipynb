```python
import pandas as pd
import matplotlib.pyplot as plt

grants = pd.read_csv('C:/Users/nswat/Documents/DataCamp/Datasets/4JoiningDataWPandasDatasets/Small_Business_Grant_AgreementsNew.csv')
licenses = pd.read_csv('C:/Users/nswat/Documents/DataCamp/Datasets/4JoiningDataWPandasDatasets/licenses.csv')
wards = pd.read_csv('C:/Users/nswat/Documents/DataCamp/Datasets/4JoiningDataWPandasDatasets/wards.csv')

grants_licenses_ward = grants.merge(licenses, on=['address', 'zip'])\
                        .merge(wards, on='ward', suffixes=('_bus', '_ward'))

grants_licenses_ward.groupby('ward').agg('sum').plot(kind='bar', y='grant')
plt.show()
```
